# What is NumPy and how is it important in Machine learning?

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

And since machine learning is all about turning data into numbers and then figuring out the patterns, NumPy often comes into play.

# Why numPy
1. **Efficient array operations**: NumPy provides highly optimized implementations of array operations, allowing for efficient computations on large datasets. It leverages low-level implementations written in C or Fortran, making it significantly faster than traditional Python lists.

2. **Multi-dimensional array representation**: Machine learning algorithms often deal with multi-dimensional data, such as images, audio, and time series. NumPy's ndarray (N-dimensional array) data structure provides a convenient and efficient way to represent and manipulate such data. It enables easy indexing, slicing, reshaping, and broadcasting operations.

3. **Vectorized computations**: NumPy allows for vectorized computations, which means you can perform operations on entire arrays or matrices at once, without the need for explicit loops. This vectorized approach is much faster and more concise than writing explicit loops in Python. It helps in improving code readability and performance.

4. **Integration with other libraries**: NumPy serves as a foundation for many other libraries and frameworks in the Python scientific computing ecosystem, including Pandas, SciPy, Scikit-learn, and TensorFlow. These libraries often expect inputs in the form of NumPy arrays and use NumPy's functionalities extensively for data manipulation, mathematical operations, and numerical computations.

5. **Mathematical functions**: NumPy provides a wide range of mathematical functions, including linear algebra, Fourier transforms, random number generation, and more. These functions are essential for implementing and optimizing machine learning algorithms.

6. **Memory efficiency**: NumPy arrays are memory-efficient compared to Python lists. They store data in a contiguous block of memory, allowing for efficient memory allocation and utilization. This aspect is especially important when working with large datasets or training deep learning models that require significant memory resources.

## What is array and ndarray?
An array is a central data structure of the NumPy library. An array is a grid of values and it contains information about the raw data, how to locate an element, and how to interpret an element. It has a grid of elements that can be indexed in various ways. The elements are all of the same type, referred to as the array dtype.

An array can be indexed by a tuple of nonnegative integers, by booleans, by another array, or by integers. The rank of the array is the number of dimensions. The shape of the array is a tuple of integers giving the size of the array along each dimension.

In NumPy, `ndarray` stands for "N-dimensional array." It is the primary data structure provided by NumPy and serves as the foundation for most numerical computations in the library. The `ndarray` is a multi-dimensional, homogeneous array object that efficiently stores and operates on large arrays and matrices.

### Here are the key characteristics and properties of the `ndarray`:

1. **N-dimensional**: The `ndarray` can have any number of dimensions. It is commonly used to represent 1-D arrays (vectors), 2-D arrays (matrices), and higher-dimensional arrays. Each dimension is referred to as an "axis."

2. **Homogeneous data type**: All elements within an `ndarray` must have the same data type. This ensures efficient memory storage and enables optimized mathematical operations.

3. **Fixed size**: The size (shape) of an `ndarray` is fixed upon creation and cannot be changed. However, you can reshape an array to have a different shape without changing the underlying data.

4. **Efficient storage**: The elements of an `ndarray` are stored in a contiguous block of memory, allowing for efficient memory allocation and access. This enables efficient processing of large datasets and seamless integration with low-level languages like C and Fortran.

5. **Fast and vectorized operations**: NumPy provides a wide range of functions and operators that operate on `ndarray`s in a vectorized manner. These operations allow for efficient computations on entire arrays or subarrays, eliminating the need for explicit loops.

6. **Indexing and slicing**: The `ndarray` supports various indexing and slicing techniques, similar to Python lists. You can access individual elements, subarrays, or specific axes using integer or boolean indexing.

## Broadcasting
### What is broadcasting?

Broadcasting is a feature of NumPy which performs an operation across multiple dimensions of data without replicating the data. This saves time and space. For example, if you have a 3x3 array (A) and want to add a 1x3 array (B), NumPy will add the row of (B) to every row of (A).

### Rules of Broadcasting

* If the two arrays differ in their number of dimensions, the shape of the one with fewer dimensions is padded with ones on its leading (left) side.
* If the shape of the two arrays does not match in any dimension, the array with shape equal to 1 in that dimension is stretched to match the other shape.
* If in any dimension the sizes disagree and neither is equal to 1, an error is raised.

***The broadcasting rule:*** In order to broadcast, the size of the trailing axes for both arrays in an operation must be either the same size or one of them must be one.
